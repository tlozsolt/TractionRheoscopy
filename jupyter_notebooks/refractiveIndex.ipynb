```python
def massFraction(ri_dataRow):
    master = ri_dataRow['master']
    pdm1922 = ri_dataRow['PDM-1922']
    pdm0821 = ri_dataRow['PDM-0821']
    
    total = master + pdm1922 + pdm0821
    total_pdm1922 = pdm1922 + (1.0067)/(1.0667 + 4.0954)*master
    return total_pdm1922/total

ri_data['mass fraction PDM-1922'] = ri_data.apply(massFraction,axis=1)
```

## Compute liquid filler fraction

```python
gelComp_dict = {'b1': 0.1667, 'b2': 0.1988, 'c':0.0144}
def liquidFiller(lmbda, b1, b2, c, reactiveWt_dict):
    b1_wt = reactiveWt_dict['b1']
    b2_wt = reactiveWt_dict['b2']
    c_wt = reactiveWt_dict['c']
    return ((1-b1_wt)*b1 + (1-b2_wt)*b2 + (1-c_wt)*c + lmbda)/(b1 + b2 + c + lmbda)
```

```python
liquidFiller(5,1,1,1, gelComp_dict)
```
