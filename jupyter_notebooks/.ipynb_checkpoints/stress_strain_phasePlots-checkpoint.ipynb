### Stress strain phase plot 

The purpose of this notebook is to show joint variation plot of stress vs. strain.

What do I mean by this? I will explain with a few simple examples of idealized stress/strain curves and their representation with joint varation or phase plots

A *linear* stress vs. strain curve would yield plots of a single point on $\delta\epsilon$ vs. $\delta\sigma$
with "error" bars representing the spread in measured stress and strain. Depending on the time window used to define the strain, the point would be located at different positions, but the vector pointing from the origin to the point would have the same angle.

A *piecewise continuous stepped* stress-strain curve would have two points on the joint plot corresponding, or more precisely, the time series would show alternating density at one of two points: $(\delta\epsilon,0)$ or $(0,\delta\sigma)$. The magntiude and angle of the "linear" approximation would be a time weighted average of the two points.

## Process
Steps to create recurrence plots 
- compute stress and strain at a given time interval, possibly a fixed time 
- set up some spatial bins in xy for coarse graining

  - These bins should be referenced to the **current** configuration of the gel as the majority of the applied displacement is carried 
    by the gel. 
  - [+] interpolate the gel to get tractions finely resolved enough for the coarse graining of the sediment
          - I think this is a better option as tracer spacing is not fine enough in xy and would be a limiting factor down
            the line in determining the spatial correlation
          - I am going to check to see if the depth dependence of the strain is significant or if it can be safely ignored
            and just included in the average. 
          - The stresses should be reported at the *current* configuration. 
  - [-] coarse grain the sediment to get sufficient particles in each bin 
         - I dont think this is the best approach as the ''optimal'' bin sizes for coarse graining in sediment and gel
           are going to be very different as a direct result of the vastly different densities of particles. 
     
- bin the sediment particles (into those bins or into optimal bins for sediment) and compute descriptive statistics in each bin
- plot joint plot of stress and strain increment over time interal
- repeat for all times (ie move the window)
- vary time interval

## 1) Interpolation

### 1a) Is strain height dependent?

- if no, then interpolate in xy regardless of height
- if yes, look into the height dependence as a consequence of st venant.
- the strain is weakly height dependent where the height dependence is an oders of magnitude smaller than the strain the gel.
- xy variation are also an order of magnitude smaller that average, but about a factor of two larger than height variations. 
- it is very difficult to rule optical aberrations affect the strain measurment. 
- it also plainly obvious that normal, out of plane motions cannot be discounted.
