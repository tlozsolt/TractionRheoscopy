### Position data
- frame refers to the stack number, ie time, during the experiment. Shear post start moving at the begining of the acquistion of frame 1 and is stationary during all of frame 0
- x (um, imageStack) is the x component in units of microns in the imageStack coordinate system (see next point) 
- imageStack coordinate have origin at top left corner of the image, with x to the right and y down...treat the image as an array and the x coordinate is scanned fastest. Furthermore, the z=0 position is somewhere in the gel and hence all particles will have signficant z-offset of 20 um or so.
- x_std is the uncertainty in the located position in pixels after gaussian refinement using trackpy. The pixel is 0.115 um/px in x and y and 0.150um/px in z, however there is some magnification errors and so this is just an average value over the image.
- if the x_std is NAN, then the gaussian refinement failed and the particle position was found using centroid fitting.

### Trajectory format

- (0,3) means the strain measured over reference configuration of frame 0 and current configuration of frame 3
- The components such thgat shear direction is x direction, so shear is exz
- "nnb count" is the number of nearest neighbors for that particular particle.
