```python
# create a binary search tree
from scipy.spatial import cKDTree
```

```python
keyList = ['{} (um, imageStack)'.format(x) for x in ['x', 'y','z']]
aidanPath_stem = '/Users/zsolt/Colloid/DATA/tfrGel10212018x/tfrGel10212018A_shearRun10292018f/locations_stitch/partial_data_Aidan'
aidan_fName = 'shearRun10292018f_centerParticles_t{:02}.h5'
#aidan_fName.format(3)
```

```python
# get particle ids for 100k particles close to the center at time 0.
zCenter,yCenter,xCenter = np.array([(80+20)/2, 235/2, 235/2])
tree = cKDTree(sedPos.xs(0,level='frame')[keyList])
centerID = tree.query_ball_point([xCenter, yCenter,zCenter],1)
for t in range(0,89):
    #print(t)
    tFrame = sedPos.xs(t, level='frame')
    center = tFrame.loc[centerID][keyList]
    tree = cKDTree(tFrame[keyList])
    idx = tree.query_ball_point(center.to_numpy(),55)[0]
    #print(len(idx))
    # Reminder: you have to use iloc, not loc when indexing between tree and dataFrame. 
    tFrame.iloc[idx].to_hdf(aidanPath_stem + '/' + aidan_fName.format(t),'t{:02}'.format(t),mode='w')
    #print(center)
```
